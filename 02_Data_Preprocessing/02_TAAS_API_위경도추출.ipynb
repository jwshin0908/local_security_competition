### 경도, 위도 변환

```python
location = df[['x_crdnt', 'y_crdnt']]
location
```

```python
from pyproj import Proj, transform
 
epsg5179= Proj(init="epsg:5179")
wgs84=Proj(init='epsg:4326')
```

```python
location['transformed'] = location.apply(lambda row : transform(epsg5179, wgs84, row['x_crdnt'], row['y_crdnt']), axis = 1)

print(location)
```

```python
location['x'] = location['transformed'].apply(lambda x : x[0])
location['y'] = location['transformed'].apply(lambda x : x[1])

location = location[['x', 'y']]
```
